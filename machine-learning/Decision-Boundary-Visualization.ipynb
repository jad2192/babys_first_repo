```python
# Imports
import numpy as np
import pandas as pd
import matplotlib as mpl
from matplotlib import pyplot as plt
from scipy import stats
import sklearn
from sklearn import datasets

%matplotlib inline
plt.style.use('seaborn-whitegrid')
```

```python
#Create a toy data set of 4 labeled classes

X,Y = datasets.make_blobs(n_samples=500,centers=4,cluster_std=2.4,random_state=8)
plt.figure(figsize=(10,10))
plt.scatter(X[:,0],X[:,1],c=Y,s=50, cmap='jet')
```

```python
# A function which takes a ML classifier and a set of labeled 2D data and returns a visualization of the
# decision boundary of the fitted classifier.

def visualize_boundary(clf, X, Y, x_lim=None, y_lim=None):
    '''Input a (scikit-learn) classifier and data (with 2 attributes) and this function will automatically
       plot a scatter of the data with the classifier's decision boundary'''
    
    clf.fit(X,Y)
    
    if x_lim == None:
        
        x_lim = (X[:,0].min() - 0.1, X[:,0].max() + 0.2)
    
    if y_lim == None:
        
        y_lim = (X[:,1].min() - 0.1, X[:,1].max() + 0.2)
    
    xx, yy = np.meshgrid(np.arange(x_lim[0],x_lim[1],0.1),np.arange(y_lim[0],y_lim[1],0.1))
    Z = clf.predict(np.c_[xx.ravel(),yy.ravel()]).reshape(xx.shape)
    
    plt.figure(figsize=(10,10))
    plt.scatter(X[:,0],X[:,1],c=Y,cmap='jet_r',s=50)
    plt.contourf(xx,yy,Z,cmap='rainbow',alpha=0.16)
    
```

```python
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
```

```python
# Introducing various sklearn ML classifiers: support vector, logistic regression, Gaussian naive bayes,
# decision tree, random forest and multi-layer perceptron with 2 hidden layers.

clf_svm = SVC(C=0.1, kernel='linear')
clf_logit = LogisticRegression()
clf_nb = GaussianNB()
clf_t = DecisionTreeClassifier(max_depth=2,criterion='entropy')
clf_rf = RandomForestClassifier(n_estimators=100)
clf_nn = MLPClassifier(hidden_layer_sizes=(100,100))
```

```python
# Decision boundary for suppor vector classifier with linear kernel

visualize_boundary(clf_svm, X, Y)
```

```python
# Decision boundary of logistic regression classifier

visualize_boundary(clf_logit, X, Y)
```

```python
# Decision boundary of Gaussian naive Bayes classifier

visualize_boundary(clf_nb, X, Y)
```

```python
# Decision boundary for Decision Tree Classifier

visualize_boundary(clf_t, X, Y)
```

```python
#Decision boundary of random forest classifier with 100 trees

visualize_boundary(clf_rf,X,Y)
```

```python
# Decision boundary for MLP classifier 

visualize_boundary(clf_nn, X, Y)
```
